#### 데이터의 정형화
- item_name의 values를 모두 대(소)문자로 변경
    - 문자열 내부함수 중 모든 문자를 대문자로 변경: upper()
- item_name의 values에 존재하는 문자 사이의 공백을 제거
    - 문자열 내부함수 중 문자열을 치환하는 함수: replace()
        - 문자열 사이의 데이터 치환
    - 문자열 내부함수 중 공백을 제거하는 함수: strip()
        - strip() 문자열 좌lstrip, 우rstrip 공백 제거
        - 본 데이터에서는 사용 불가

```python
# item_list를 기준으로 반복 실행하는 조건문 생성
for i in item_list:
    # item_list는 앞서 생성해둔 중복이 없는 상품명 리스트 A~Z
    item = uriage["item_name"] == i
    # item = i(A~Z) 값이 대입 

    # 상품 A~Z 각각 결측치가 아닌 item_price의 평균을 추출
    price_mean = uriage.loc[
        item & ~flag,
        "item_price"
    ].mean()

    # 상품 A~Z 각각 결측치인 데이터에 상기 price_mean을 대입
    uriage.loc[
        item & flag,
        "item_price"
    ]=price_mean
```
