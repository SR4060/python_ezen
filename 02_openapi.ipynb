```python
import requests
```

```python
# local server에 있는 데이터 api에 요청
url="http://127.0.0.1:5000/data"

# 서버에게 보낼 데이터를 dict 형태로 생성 
parameter={
    'service_id':'test10',
    'service_pass':'1111'
}

# 서버에게 요청을 보낸다. (데이터를 실어서 보냄)
reponse = requests.get(url, params=parameter)
```
