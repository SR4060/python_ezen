### bs4 라이브러리
- bs4 라이브러리 안에 있는 BeautifulSoup Class 사용
- html로 이루어진 문자열 데이터를 parsing 작업하여 데이터를 쉽게 추출하기 위한 Class
    - parsing
        - Class에 독립변수 값을 설정
        - 데이터의 타입을 변경
- html 데이터를 TAG를 기준으로 데이터를 추출하는 기능
- 웹의 구조를 어느정도 인지하고 사용하게 되면 쉽게 사용이 가능

```python
from bs4 import BeautifulSoup as bs
```

```python
# htmla 데이터를 bs Class 에 대입하여 데이터의 타입을 변경
soup = bs(html_data, 'html.parser')
```

- 태그의 이름을 기준으로 검색
    - soup.태그명: 해당 태그명의 첫번째 태그를 출력
    - soup.태그명.string: 첫번째 태그명의 문자열(content)을 출력
        - ex) `<p>test</p>` --> test
    - soup.태그명['속성명']: 첫번째 태그에서 해당 속성의 값을 출력
        - ex) `<a href='https://www.google.com'>Google</a>` --> https://www.google.com

- BeautifulSoup 안에 함수
    - find()
        - html 데이터에서 특정 태그의 첫번째 정보를 출력
        - find(속성명 = 속성값): 태그들 중 해당 속성명과 속성값을 가진 첫번째 태그의 정보를 출력
        - 결과값의 타입이 TAG의 형태로 출력: find() 함수 이용 가능
    - find_all()
        - html 데이터에서 특정 태그를 모두 출력
        - limit 매개변수: 출력할 태그의 개수를 설정
        - 결과값의 타입이 ResultSet 형태로 출력: find() 함수 이용 불가

```python
# 네이버 파이넨스 사이트에 요청
finance_url='https://finance.naver.com/'

# requests 라이브러리를 이용하여 요청
finance_res=requests.get(finance_url)
```

```python
# bs를 이요하여 데이터를 파싱
finance_soup=bs(finance_res.text, 'html.parser')
```

```python
finance_soup
```

```python
# table 태그 중 class의 값이 tbl_home인 태그의 개수를 확인
len(
    finance_soup.find_all('table', attrs={
        'class':'tbl_home'
    })
)
```

```python
# table 에서 class가 tbl_home의 개수가 너무 많다
# 부모 태그인 div에 class가 section_sise_top인 태그의 개수를 확인

len(
    finance_soup.find_all('div', attrs={
        'class':'section_sise_top'
    })
)
```

```python
# 태그의 개수가 1개인 것을 확인
# 전체 html에서 div가 section_sise_top인 태그로 라인의 수를 줄인다.
section_data=finance_soup.find('div', attrs={
    'class':'section_sise_top'
})
```

```python
section_data
```

```python
# section_data에서 table태그 중 class가 tbl_home인 태그의 개수를 확인
len(
    section_data.find_all(
        'table', attrs={
            'class':'tbl_home'
        }
    )
)
```

```python
# table 4개의 정보를 모두 저장
tables=section_data.find_all('table')
```

```python
# table 중 첫번째 table의 정보를 확인
table1=tables[0]
```

```python
table1
```

```python
# table1에서 thead의 값들을 출력하여 DataFrame에 columns의 값을 사용
thead_data=table1.find('thead')
```

```python
# thead_data에서 th태그를 모두 출력
th_list=thead_data.find_all('th')
```

```python
# th_list에서 문자들만 출력하여 새로운 리스트에 대입
# 리스트 형태의 데이터에서 각 원소들에 특정 행동을 하는 함수 -> map()

cols=list(
    map(
        lambda x: x.get_text(),
        th_list
    )
)
```

```python
# table1에서 tbody부분 데이터를 출력
tbody_data=table1.find('tbody')
```

```python
# tbody_data에서 tr들을 모두 출력
tr_list=tbody_data.find_all('tr')
```

```python
# tr_list에서 첫번째 tr만 추출
tr_data=tr_list[0]
```

```python
# tr_data에서 th태그, td태그를 출력
th_list=tr_data.find_all('th')
td_list=tr_data.find_all('td')
value_data=th_list+td_list
```

```python
[ value.get_text().strip() for value in value_data ]
```

```python
# 위의 value 작업들을 반복 실행
# tr_list의 개수만큼

values=[]

for tr_data in tr_list:
    
    th_list=tr_data.find_all('th')
    td_list=tr_data.find_all('td')
    value_data=th_list+td_list
    val = [ value.get_text().strip() for value in value_data ]
    values.append(val)
values
```

```python
import pandas as pd
```

```python
# cols, values를 이용하여 DataFrame 생성
df=pd.DataFrame(values, columns=cols)
```

```python
# sqlalchemy 라이브러리 안에 있는 create_engine 기능을 로드
from sqlalchemy import create_engine
```

```python
# 서버의 정보를 저장
user='root'
passworld='1234'
host='localhost'
port=3306
database='ezen'
# sqlalchemy를 이용하여 데이터베이스 서버와 연결
connection_url=f'mysql+pymysql://{user}:{passworld}@{host}:{port}/{database}'
engine=create_engine(connection_url)
```

```python
# dfs에 있는 첫번째 데이터프레임을 데이터베이스에 대입
# to_sql( 테이블명, con=데이터베이스정보, index=인덱스저장유무, if_exists= 'replace' | 'append' ) 
dfs[0].to_sql('trade_top', con=engine, index=False, if_exists='append')
```
